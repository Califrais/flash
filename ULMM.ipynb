```python
%matplotlib inline

import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = sm.datasets.get_rdataset('dietox', 'geepack').data
md = smf.mixedlm("Weight ~ Time", data, groups=data["Pig"])
mdf = md.fit()
print(mdf.summary())
```

```python
mdf.params
```

```python
mdf.resid.values.var()
```

```python
md = smf.mixedlm("Weight ~ Time + Start", data, groups=data["Pig"], re_formula="~Start")
mdf = md.fit()
print(mdf.summary())
```
